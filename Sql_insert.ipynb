```python
import mysql.connector
from datetime import datetime 
import random
import string
import time
from uuid import uuid4


# Database connection parameters
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'root',
    'database': 'new'
}


def insert_records(conn):
        cursor = conn.cursor()
        name_ran=str(uuid4())
        created_at=str(datetime.now())
        sql = "INSERT INTO kafka (name_ran,created_at) VALUES (%s, %s)"
        val = (name_ran,created_at)
        cursor.execute(sql, val)
        conn.commit()
        print("Record inserted successfully:", val)
        cursor.close()
    
i=0
while True:
    
    try:
        connection=mysql.connector.connect(**db_config)
        insert_records(connection)
        connection.close()
        print("Connection closed")
        # Wait for 5 seconds before next iteration
        time.sleep(5)
    except Exception as e:
        print("Error:", e)
        # If an error occurs, wait for 5 seconds before retrying
        time.sleep(5)
#     i+=1
```
